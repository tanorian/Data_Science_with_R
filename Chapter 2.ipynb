Matrices are like data frames but all values in the object must have the same data type.  Use single brackets to access their values.

"Recycling" is when R starts over at the beginning of a vector when a vector operation needs more values than a vector has.  This can cause some serious problems in your programs if you're unaware of it.  No errors are thrown.  Here is a simple example:

```r
x <- c(1, 2, 3, 4)
y <- c(10, 20, 30, 40, 50, 60, 70, 80, 90)
x + y
```

What is the per 100,000 murder rate for each state?

```r
murder_rate <- 100000*murders$total/murders$population
```

Create a vector of murder rates comparable to or less than that of Italy (0.71 per 100,000 people).

```r
ind <- murder_rate <= 0.71  # a Boolean vector
```

```r
murders$state[ind]
```

```r
ind  # This is called subsetting with logicals
```

The '&' operator is a logical AND:

```r
#  List states that are in the West AND have a murder rate per 100,000 equal to 1 or less.
west <- murders$region == 'West'
safe <- murder_rate <= 1
ind <- safe & west
murders$state[ind]
```

What is the homicide rate in California?

```r
ind <- which(murders$state == 'California')
murder_rate[ind]
```

Use the match() function to return the vector indexes for values that match a given list:

```r
ind <- match(c('Texas', 'New York', 'Florida'), murders$state)
murder_rate[ind]
```

You can test to see if values in one vector exist in a second vector using the %in% operator.

```r
c("Boston", "California", "Philadelphia")%in%murders$state
```

```r
match(c("Florida", "California", "Ohio"),murders$state)
which(murders$state%in%c("Florida", "California", "Ohio"))
```
