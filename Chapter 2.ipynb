Data frames are types of lists...  Another way to access members of lists or data frames is using the double brackets.  If the list does not have names for columns then you must use the double brackets.

Matrices are like data frames but all values in the object must have the same data type.  Use single brackets to access their values.

```r
mat <- matrix(1:12, 4, 3) # arguments are (Values, # of rows, # of columns)
mat[2,3]  # get a specific value at (row, column)
mat[3, ]  # get the entire third row
mat[ ,2]  # get the entire second column
mat[ ,2:3]# get all rows from the second and third column
mat[1:2,2:3]# get the first and second rows from the second and third column
```

You can convert matrices to data frames with the as.data.frame() function:

```r
as.data.frame(mat)  # The "V" stands for Vector
```

You can also use single brackets to access values in a data frame:

```r
murders[20:24, 2:3]
```

```r
vec <- c("1", "2", "3", "4", "5")
class(vec)
class(as.integer(vec))
y <- as.numeric(vec)
class(y)
z <- as.character(y)
class(z)
```

SORTING

The sort() function displays the values of a vector in order.

```r
library(dslabs)
data(murders)
sort(murders$total)
```

But this say what states those values are associated with.  Use the order() function to get this.  sort() returns the values in order; order() returns the indexes of the sorted values.

```r
x <- c(31,4,15,92,65)
sort(x)
index <- order(x)
index
x[index]
```

Now, looking at the murder data and using the order() function:

```r
ind <- order(murders$total)
murders$abb[ind]
```

Want to know which state had the highest number of murders?  Use the max() and which.max() functions.

```r
max(murders$total)
ind_max <- which.max(murders$total)
murders$state[ind_max]
```

For the fewest, use the min() and which.min() functions.

```r
min(murders$total)
ind_min <- which.min(murders$total)
murders$state[ind_min]
murders$abb[ind]
```

Another similar function is rank() - it returns a vector with the rank (smallest to largest) of each value in the original vector.

```r
x <- c(31,4,15,92,65)
x
sort(x)
order(x)
rank(x)
x
```

"Recycling" is when R starts over at the beginning of a vector when a vector operation needs more values than a vector has.  This can cause some serious problems in your programs if you're unaware of it.  No errors are thrown.  Here is a simple example:

```r
x <- c(1, 2, 3, 4)
y <- c(10, 20, 30, 40, 50, 60, 70, 80, 90)
x + y
```
