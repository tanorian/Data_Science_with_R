Data types:

You can determine the data type of a variable with the class() function:

```r
class(a)
```

A data frame is an object that stores values in rows and columns.

```r
data(murders) # loads the data set
class(murders)
```

To see the structure of an object, use the str() function:

```r
str(a)
str(murders)
```

Use the head() function to see the top six lines in the data frame:

```r
head(murders)
```

Access the values in a column with the "$" symbol.

```r
murders$population
murders$state
#  The order in which they appear in the data frame is preserved!
```

The names() function lists the names of the columns in a data frame.

```r
names(murders)
```

In practice, all values are vectors (multiple values which all have the same data type).  The length of the vector displays how many values the vector contains.

```r
length(murders$population)
class(murders$population)
class(murders$state)
length(a)
```

Integers are usually stored as numerics unless specifically called an integer with the "L" tag:

```r
a <- 4
class(a)
a <- 4L
class(a)
```

Factors are vectors with only a small number of allowed values.  Use the levels() function to see the values.

```r
class(murders$region)
levels(murders$region)
```

Order the levels by total murders in region

```r
region <- murders$region
value <- murders$total
region <- reorder(region, value, FUN=sum)
levels(region)
```

Data frames are types of lists...  Another way to access members of lists or data frames is using the double brackets.  If the list does not have names for columns then you must use the double brackets.

```r
murders[["region"]]
```

```r
murders[[3]]
```

Matrices are like data frames but all values in the object must have the same data type.  Use single brackets to access their values.

```r
mat <- matrix(1:12, 4, 3) # arguments are (Values, # of rows, # of columns)
mat[2,3]  # get a specific value at (row, column)
mat[3, ]  # get the entire third row
mat[ ,2]  # get the entire second column
mat[ ,2:3]# get all rows from the second and third column
mat[1:2,2:3]# get the first and second rows from the second and third column
```

You can convert matrices to data frames with the as.data.frame() function:

```r
as.data.frame(mat)  # The "V" stands for Vector
```

You can also use single brackets to access values in a data frame:

```r
murders[20:24, 2:3]
```

In R, the most basic object is the vector.  Even single values are just simple vectors.  You create vectors of more than one value with the c() function - that stands for "concatenate".  All values in the vector have the same data type.

```r
codes <- c(380, 124, 818, "try", 'this') # type of quote doesn't matter, single or double
class(codes)
codes
```

You can name the values in the vector too:

```r
country <- c("italy"=308, "canada"=124, "egypt"=816)
names(country)
country[3]
country["canada"]
country <- c(italy=308, canada=124, egypt=816)
names(country)
country[3]
country[canada]  # But you still need quotes to use the value names! 
country_b <- c(italy <- 308, canada <- 124, egypt <- 816) # BUT THIS DOESN'T NAME THE VALUES!
names(country_b)
egypt 
country_b[1]
```

The seq() function creates a vector.  Arguments are (start_value, end_value, step value default is 1)

```r
seq(3, 14)
seq(2, 11, 2)  # never makes it to 11 due to step value.
seq(1,2, 0.1)
```

You can 'subselect' values from vectors like with matrices.

```r
codes
codes[2:3]
```

```r
codes[c(1,3,5)]
country["canada"]
country[c("egypt", "canada")]
country[c(1,2)]
```
